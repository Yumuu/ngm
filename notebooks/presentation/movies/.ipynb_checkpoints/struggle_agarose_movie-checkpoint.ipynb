```python
import numpy as np
from scipy.interpolate import interp1d
import shutil
import matplotlib
matplotlib.rc('font', family='arial') 
import matplotlib.pyplot as plt
import dask.array as da
import flika as flk
from glob import glob
flk.start_flika()
from flika.window import Window as flw
from dask.array.image import imread as daim
from moviepy.editor import ImageSequenceClip, VideoClip, concatenate_videoclips
from moviepy.video.io.bindings import mplfig_to_npimage
from skimage.exposure import rescale_intensity

from json import load
from skimage.io import imread
from fishtrack.measurements import tail_angle
from scipy.ndimage.filters import median_filter
from fish.image.vol import dff
```

```python
base_dir = '/nrs/ahrens/davis/data/epi/20170621/6dpf_cy221xcy221_f2_spon_1/'
exp_name = base_dir.split('/')[-2]
```

```python
dset = daim(base_dir + 'Pos0/*.tif')
with open(glob(base_dir + '*.json')[0]) as f:
    params = load(f)
```

```python
reg = imread(glob(base_dir +'*aligned.tif')[0])
```

```python
ang = tail_angle(reg, -np.array(params['fish_bbox'][:2])[::-1], 100)[0]
vel = rescale_intensity(np.abs(np.diff(ang)), out_range=(0,1))
```

```python
plr = np.arange(12000, 24000, 10)
raw = dset[plr].compute()
```

```python
fl_roi = 102 + np.arange(-5,3), 190 + np.arange(-5,5).reshape(-1,1)
```

```python
10 * 300
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(12,4))

axs[0].imshow(reg[-1][fl_roi], clim=(2000,3000))
dff_lims = (-.015, .12)
dff_ts = dff(median_filter(reg[:,fl_roi[0], fl_roi[1]].mean((1,2)), size=300), window=8000, percentile=50, baseline_offset=1000, downsample=10)
dff_ts = rescale_intensity(dff_ts, in_range=dff_lims, out_range=(0,1))
axs[1].plot(rescale_intensity(reg[:,fl_roi[0], fl_roi[1]].mean((1,2))[plr], out_range=(0,1)))
axs[1].plot(dff_ts[plr])
axs[1].plot(vel[plr])
#axs[1].set_xlim(0,400)
```

```python
ind = 0
data_dur = 45
reg_ = reg[plr].transpose(0,2,1)
fps = len(reg_) / data_dur
figsize=(8, 4.5)
fig = plt.figure(figsize=figsize, dpi=200, facecolor='k')
pad = .1
ax_raw = fig.add_axes([pad, .1 + pad,      .4, .7])
ax_reg = fig.add_axes([.4 + pad, .1 + pad, .4, .7])
ax_ts = fig.add_axes([pad, .05, .8, .15], facecolor='k')

window = np.arange(-600, 600)

plots = []
plots.append(ax_ts.plot(window, vel[window + plr[ind]], color='w')[0])
plots.append(ax_ts.plot(window, dff_ts[window], color='green')[0])
ax_ts.axvline(0, color='y')
ax_ts.set_ylim(0, 1)
[ax.axis('off') for ax in (ax_raw, ax_reg)]
label_font = dict(size=10)
gcamp_label = dict(x = 0, y=1.15, s='$\it{Tg(gfap:GCaMP6f)}$', transform=ax_ts.transAxes, color=(1,1,1), fontdict=label_font)
raw_label = dict(x = .1, y=.95, s='Raw image', transform=ax_raw.transAxes, color=(.8,.8,.8), fontdict=label_font)
reg_label = dict(x = .1, y=.95, s='Registered image', transform=ax_reg.transAxes, color=(.8,.8,.8), fontdict=label_font)
vel_label = dict(x = 0, y=.3, s='Angular tail velocity ', transform=ax_ts.transAxes, color='w', fontdict=label_font)
dff_label = dict(x = 0, y=.06, s='\u0394f/f', transform=ax_ts.transAxes, color='g', fontdict=label_font)
timing_label = dict(x = 0, y = -.20, s=f'{fps/(100/10):.2f}X playback speed',
                    transform=ax_ts.transAxes, 
                    fontdict=label_font, color=(.8,.8,.8))


clim = (1500, 5000)
ims = []
ims.append(ax_raw.imshow(raw[ind], cmap='inferno', clim=clim))
ims.append(ax_reg.imshow(reg_[ind], cmap='inferno', clim=clim))
ax_ts.text(**gcamp_label)
ax_raw.text(**raw_label)
ax_reg.text(**reg_label)
ax_ts.text(**vel_label)
ax_ts.text(**dff_label)
ax_ts.text(**timing_label)
def drawf(t):
    global ind
    
    if t == 0.0:
        ind = 0
    
    if ind >= reg_.shape[0]:
        ind = 0
    ims[0].set_data(raw[ind])
    ims[1].set_data(reg_[ind])
    plots[0].set_data(window, vel[window + plr[ind]])
    plots[1].set_data(window, dff_ts[window + plr[ind]])
    ind += 1
    return mplfig_to_npimage(fig)

fig_, axs_= plt.subplots(figsize=(12,12))
axs_.imshow(drawf(0))

data_clip = VideoClip(drawf, duration=data_dur)
```

```python
## make the title card 
title_dur = 4
fig_intro = plt.figure(dpi=200, figsize=np.array(figsize), facecolor='k')
ax_intro = fig_intro.add_axes((0, 0, 1, 1))
ax_intro.set_facecolor('k')
intro_text =r'''Video S5:

Calcium imaging during futility-induced passivity in a 
fish attempting to swim in viscous agarose'''
            
ax_intro.text(.15,
              .8,
              intro_text, 
              fontdict=dict(color='w', size=16), 
              transform=ax_intro.transAxes, 
              ha = 'left',
              va='top',
              multialignment='left')


def drawf_intro(t_):
    return mplfig_to_npimage(fig_intro)

title_clip = VideoClip(drawf_intro, duration=title_dur)
```

```python
_,_  = plt.subplots(figsize=(12,12))
plt.imshow(drawf_intro(0))
```

```python
vid_fname = f'/groups/ahrens/ahrenslab/davis/tmp/{exp_name}_t{plr[0]}t{plr[-1]}.mp4'
composite_clip = concatenate_videoclips([title_clip, data_clip])
ind = 0
composite_clip.write_videofile(vid_fname, audio=False, codec='h264', fps=fps, ffmpeg_params=['-pix_fmt', 'yuv420p'])
shutil.copyfile(vid_fname, '/groups/ahrens/ahrenslab/davis/tmp/Video_S5.mp4')
```
