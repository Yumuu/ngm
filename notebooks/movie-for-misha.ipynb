```python
import numpy as np
import moviepy
from joblib import Parallel, delayed
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.transform import resize
%matplotlib inline
```

```python
from glob import glob
fnames = sorted(glob('/groups/ahrens/ahrenslab/davis/data/spim/proc/20160607/5dpf_cy171xcy221_f1_omr_1_20160607_183009/projs_z/*.tif'))
```

```python
%%time
data = np.array(Parallel(n_jobs=16)(delayed(imread)(f) for f in fnames[:1000]))
```

```python
data.shape
```

```python
mx = data.max(1)
cropped = mx[:,mx.shape[1]//2:,:]
```

```python
%%time
resize_fun = lambda v: resize(v, output_shape=(v.shape[0] * 2, v.shape[1]))
stretched = np.array([resize_fun(cropped[:,y]) for y in range(cropped.shape[1])]);
stretched = np.transpose(stretched, (1,0,2)).astype('int16')
```

```python
# save cropped, max projected, stretched array to disk
from skimage.io import imsave
out_dir = '/groups/ahrens/ahrenslab/projects/ngm/presentation/resource/5dpf_cy171xcy221_f1_omr_1_20160607_183009_t0-1000_interp.tif'
imsave(out_dir, stretched)
```
